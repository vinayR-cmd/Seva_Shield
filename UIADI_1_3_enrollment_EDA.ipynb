1Ô∏è‚É£ Child enrolment (age 0‚Äì5) ‚Üí HIGHEST risk

Why?

Physical growth is rapid

Fingerprints & iris change

Biometric re-capture is mandatory later

Highest chance of mismatch

üìå Hence highest weight ‚Üí 0.6

üßí 2Ô∏è‚É£ Teen enrolment (age 5‚Äì17) ‚Üí MODERATE risk

Why?

Growth continues but slower

Biometrics more stable than infants

Still requires periodic updates

üìå Hence medium weight ‚Üí 0.3

üë®‚Äçü¶≥ 3Ô∏è‚É£ Adult enrolment (18+) ‚Üí LOWEST risk

Why?

Biometrics are mostly stable

Updates mostly due to wear & tear (labour, aging)

Less frequent than children

üìå Hence lowest weight ‚Üí 0.1

üìê Why weighted sum? (Mathematical reasoning)

We want:

A continuous score (0‚Äì1)

Easy interpretability

Explainable to non-technical people

```python
# Risk score between 0 and 1
monthly_df['risk_score'] = (
    0.6 * monthly_df['child_enrolment_ratio'] +
    0.3 * monthly_df['teen_enrolment_ratio'] +
    0.1 * monthly_df['adult_enrolment_ratio']
)
```

calculating tjhe percentage of the child , teen and adult enroll s in each state

```python
# child_%  = sum(age_0_5) / sum(total) * 100
# teen_%   = sum(age_5_17) / sum(total) * 100
# adult_%  = sum(age_18_greater) / sum(total) * 100
```

```python
state_age_pct = (
    df.groupby('state')
      .agg(
          child_enrollments=('age_0_5','sum'),
          teen_enrollments=('age_5_17','sum'),
          adult_enrollments=('age_18_greater','sum'),
          total_enrollments=('total_enrollments','sum')
      )
)

state_age_pct['child_pct'] = (state_age_pct['child_enrollments'] / state_age_pct['total_enrollments']) * 100
state_age_pct['teen_pct']  = (state_age_pct['teen_enrollments']  / state_age_pct['total_enrollments']) * 100
state_age_pct['adult_pct'] = (state_age_pct['adult_enrollments'] / state_age_pct['total_enrollments']) * 100

state_age_pct = state_age_pct.reset_index()
state_age_pct
```

```python
import seaborn as sns

heat = state_age_pct.set_index('state')[['child_pct','teen_pct','adult_pct']]

plt.figure(figsize=(10,6))
sns.heatmap(heat, annot=True, fmt=".1f", cmap='winter', linewidth=0.6)
plt.title('Enrollment Percentage Heatmap by State')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18,6), sharey=True)

# Child %
sns.barplot(
    data=state_age_pct.sort_values('child_pct', ascending=False),
    x='child_pct',
    y='state',
    ax=axes[0]
)
axes[0].set_title('Child Enrollment %')
axes[0].set_xlabel('Percentage (%)')
axes[0].set_ylabel('State')

# Teen %
sns.barplot(
    data=state_age_pct.sort_values('teen_pct', ascending=False),
    x='teen_pct',
    y='state',
    ax=axes[1]
)
axes[1].set_title('Teen Enrollment %')
axes[1].set_xlabel('Percentage (%)')
axes[1].set_ylabel('')

# Adult %
sns.barplot(
    data=state_age_pct.sort_values('adult_pct', ascending=False),
    x='adult_pct',
    y='state',
    ax=axes[2]
)
axes[2].set_title('Adult Enrollment %')
axes[2].set_xlabel('Percentage (%)')
axes[2].set_ylabel('')

plt.tight_layout()
plt.show()
```

Based on the visualizations generated, here are the key insights into Aadhar enrollment trends:

### 1. Overall Monthly Enrollment Trend
- The line plot of overall monthly enrollments showed significant fluctuations throughout the year. There was a notable peak in September and November, suggesting higher enrollment activity during these months. Conversely, March, April, May, June, and July showed lower enrollment figures.

### 3. Daily Enrollment Patterns
- The daily enrollment patterns bar chart revealed that **Tuesday** had the highest total enrollments, followed by Monday and Thursday. Weekends (Saturday and Sunday) generally showed lower enrollment numbers, which is an expected trend given office hours and public accessibility.

### 4. Risk Score Distribution by State
- The box plot of risk score distribution across states showed varying levels of risk among states. Some states exhibited a wider interquartile range and more outliers, indicating greater variability in their monthly enrollment age-group distributions. Meghalaya and Nagaland appear to have higher median risk scores, while states like Himachal Pradesh and Lakshadweep show lower risk scores with less variability.

- **Child Enrollment Ratio vs. Risk Score:** There appears to be a strong positive correlation between the child enrollment ratio and the risk score. - **Teen Enrollment Ratio vs. Risk Score:** The scatter plot for teen enrollment ratio shows a moderate positive correlation with the risk score. Higher teen enrollment ratios also contribute to higher risk scores, though perhaps not as directly or strongly as child enrollments.
